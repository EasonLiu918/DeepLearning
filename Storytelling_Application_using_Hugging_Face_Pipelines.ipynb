```python
from transformers import pipeline
from PIL import Image

def img2text(image_path):
    #  Load Hugging Face Models | 載入 Hugging Face 的模型
    model = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")
    # Image to Text (Generate Caption) | 圖片 → 文字（生成描述）
    text = model(image_path)[0]["generated_text"]
    return text
# Run a test | 運行测试
image_path = "Cooking.jpg"
caption = img2text(image_path)
print("Image Description:", caption)
```

#Text Generation Modeling with Hugging Face

```python
from transformers import pipeline

#Select the model on the Hugging Face | 選擇 Hugging Face 上的模型
def text2story(text):
    story_generator = pipeline("text-generation", model="facebook/opt-1.3b")
    prompt = f"Once upon a time, {text}. This is the beginning of a magical adventure..."
    story = story_generator(prompt,
                            max_length=250,  ## Make the story longer | 讓故事更長
                            do_sample=True,
                            temperature=0.8,  # Slightly lower the randomness, keep the logic | 稍微降低隨機性，保持邏輯性
                            top_p=0.9,
                            repetition_penalty=1.2,  # # Avoid repeating sentences | 避免重複句子
                            truncation=True)[0]['generated_text']
    return story

# Run a test | 運行測試
story = text2story(caption)
print("Generated Story:\n", story)

from transformers import pipeline

```

#Text-to-speech

```python
from gtts import gTTS
import IPython.display as ipd

def text2audio_gtts(story_text, filename="story.mp3"):
    tts = gTTS(text=story_text, lang="en")  # Choose english | 選擇英語
    tts.save(filename)
    return filename

# RunTTS | 運行 TTS
audio_file = text2audio_gtts(story)  # `story` 是 text2story() 生成的文本

# Play Audio | 播放音頻
ipd.Audio(audio_file)
```
